```python
from uncertainties import unumpy
from uncertainties import ufloat
from scipy import odr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print('Reading file...')
df = pd.read_excel("data.xlsx")
print('File read successfully\n\n')
```

```python
theta = np.array(df['theta'])
voltage = np.array(df['volt'])
```

```python
offsetVoltage = 0.003
```

```python
errVolt = 0.001
errTheta = 0.001
```

```python
theta = unumpy.uarray(theta, errTheta)
voltage = unumpy.uarray(voltage, errVolt) - offsetVoltage
```

```python
plt.plot(unumpy.nominal_values(voltage), unumpy.nominal_values(theta), 'o', label='Data points')
plt.errorbar(unumpy.nominal_values(voltage), unumpy.nominal_values(theta), xerr=unumpy.std_devs(voltage), yerr=unumpy.std_devs(theta), fmt='o', label='Error bars')
plt.xlabel('Voltage (V)')
plt.ylabel('Theta (rad)')
plt.title('Theta vs Voltage')
plt.legend()
plt.grid()
plt.show()
```
